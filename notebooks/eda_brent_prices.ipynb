# Brent Oil Price Exploratory Data Analysis (EDA)

Analyze historical Brent oil prices and identify major events impacting the market.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

%matplotlib inline
sns.set_theme(style="whitegrid")
```

```python
# Load cleaned data
data_path = "../data/processed/BrentOilPrices_cleaned.csv"
df = pd.read_csv(data_path, parse_dates=['Date'], index_index=False)
df.set_index('Date', inplace=True)
df.head()
```

```python
# Plot historical prices
plt.figure(figsize=(15, 6))
plt.plot(df.index, df['Price'], label='Brent Oil Price (USD/barrel)')
plt.title('Historical Brent Oil Prices (1987-2026)')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.legend()
plt.show()
```
